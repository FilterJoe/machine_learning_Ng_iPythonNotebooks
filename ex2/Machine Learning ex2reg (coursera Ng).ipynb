## Goal

The goal of this exercise is implement logistic regression WITH regularization, developing intuition about how regularization can be used to avoid overfitting or underfitting.

## Use Case

Logistic Regression is a superverised learning classifier $h_\theta(x)$ to estimate the probabilility of whether a binary output is 1 or 0 based on inputs x. This example is a typical classification problem for which logistic regression is well suited: Predict whether microchips pass qaulity assurance (QA), based on a logistic classification model established from past data.<br><br>To help better fit the data, we can create more features from each data point. However, doing so may cause overfitting of the data which makes for poor generalization, and therefore poor performance when exposed to data outside the training set. By using <b>regularization</b> to reduce the influence of the additional features, we can avoid overfitting.

## Part 1: Regularized Logistic Regression

Logistic regression is used for binary classification problems, so analog output is converted to binary as follows:<br>
If $h_\theta(x)\geq .5$, then 1<br>
If $h_\theta(x)\lt .5$, then 0<br>
While several functions can be used for g in the logistic regression hypothesis $h_\theta(x) = g(\theta^Tx)$, in this exercise we use the popular Sigmoid function:<br>
$$g(z) = \frac{1}{1+e^{-z}}$$
The logistic regression cost function <i>including regularization</i> (last term) is:

$$J(\theta) = \frac{1}{m}\sum\limits_{i=1}^{m} [-y^{(i)}log(h_\theta(x^{(i)})-(1 - y^{(i)})log(1 - h_\theta(x^{(i)}))]+\frac{\lambda}{2m}\sum\limits_{j=1}^{n}\theta^2_j$$

while the gradient of the cost <i>including regularization</i> (last term) is:

$$\frac{\partial{J(\theta)}}{\partial{\theta_j}} = \frac{1}{m}\sum\limits_{i=1}^{m} (h_\theta(x^{(i)}) - y^{(i)})x_j^{(i)}     \qquad \qquad \qquad   for \,\, j=0 $$

$$\frac{\partial{J(\theta)}}{\partial{\theta_j}} = (\frac{1}{m}\sum\limits_{i=1}^{m} (h_\theta(x^{(i)}) - y^{(i)})x_j^{(i)}) + \frac{\lambda}{m}\theta_j    \qquad    for \,\, j \geq 1 $$


```python
def map_feature(X1, X2, degree=6):
    """
    Return a new r X c 2D array with more features, comprising of X1, X2, X1.^2, X2.^2, X1*X2, X1*X2.^2, etc... where
        r is number of rows in X1
        c is number of features, which is a function of degree (i.e. if degree is 6, num_features is 28)
    Args:
        X1, X2: r X 1 2D arrays (floats) with identical shape
        degree: int that represents the highest degree polynomial
    Implementation Note:
        Much more efficient to set up array once at the beginning rather than keep creating a new array each iteration
        with something like np.hstack.
    """
    num_features = (degree + 1) * (degree + 2) / 2
    out = np.empty(shape=(X1.shape[0], num_features), dtype=float)
    k = 0
    for i in xrange(0, degree + 1):
        for j in xrange(i + 1):
            new_feature_values = X1**(i - j) * X2**j
            out[:, k] = new_feature_values[:, 0]  # column assignment
            k += 1
    return out
```

```python
X = map_feature(X[:, 0][:, np.newaxis], X[:, 1][:, np.newaxis])
# mapFeature handles adding intercept term (a column of ones) to x along vertical axis
# newaxis is needed to keep each X column 2D
```

```python
def sigmoid(z):
    """
    Return sigmoid of z
    Args:
        Z (scalar or array)
    """
    return 1.0 / (1.0 + np.exp(-z))

def h_of_theta(theta, X):
    transposed_theta = theta[:, np.newaxis]
    return sigmoid(X.dot(transposed_theta))

def cost_function_reg(theta, X, y, lamda):
    """
    Return cost of using theta for logistic regression
    Args:
        theta (1D numpy array): scipy optimization requires theta parameter to be first, and passed as 1D array
        X (m x n numpy array): first column is 1 (intercept), any number of columns to follow
        y (m x 1 numpy array): each number represents the answer to the corresponding row in X
        lamda (float): regularization parameter purposely misspelled to avoid conflict with python's lambda keyword
    """
    m = len(y) * 1.0
    h_theta = h_of_theta(theta, X)

    # must change values of 0 or 1 slightly to avoid log(0)
    tol = .00000000000000000000001  # tolerance must be no higher than .00000000000000000000001 to match octave results
    h_theta[h_theta < tol] = tol  # close-to-zero values get set to tol
    h_theta[(h_theta < 1 + tol) & (h_theta > 1 - tol)] = 1 - tol  # close-to-1 values get set to 1 - tol

    regularization_term = (float(lamda)/2) * theta**2
    cost_vector = y * np.log(h_theta) + (-y + 1) * np.log(-h_theta + 1)

    J = -sum(cost_vector)/m + sum(regularization_term[1:])/m

    return J[0]


def gradient_reg(theta, X, y, lamda):
    """
    Return gradient of theta for logistic regression
    Args:
        theta (1D numpy array): theta is first parameter and a 1D array because scipy optimization requires it
        X (m x n numpy array): first column is intercept, any number of columns to follow
        y (m x 1 numpy array): each number represents the answer to the corresponding row in X
        lamda (float): regularization parameter purposely misspelled to avoid conflict with python's lambda keyword
    """
    m = len(y)
    h_theta = h_of_theta(theta, X)
    derivative_regularization_term = float(lamda) * theta/m
    grad = (h_theta - y).T.dot(X)/m + derivative_regularization_term.T
    grad[0][0] -= derivative_regularization_term[0]  # theta_0 is the only term that does not get regularized, so back out
    
    return np.ndarray.flatten(grad)  # scipy optimization requires gradient to be a 1D array
```

```python
# Test one iteration of regularized logistic regression at initial theta

m, n = X.shape
initial_theta = np.zeros(n)
lmda = 1

cost = cost_function_reg(initial_theta, X, y, lmda)
grad = gradient_reg(initial_theta, X, y, lmda)

print 'Cost at initial theta (zeros):', cost
print '\nGradient at initial theta (zeros):\n'
print grad
```

## Part 2: Regularization and Accuracies

```python
# set options then run fmin_bfgs to obtain optimal theta (BFGS: quasi-Newton method of Broyden, Fletcher, Goldfarb, and Shanno)

initial_theta = np.zeros(n)
lmda = 1
myargs = (X, y, lmda)
options = {'full_output': True, 'maxiter': 400}

optimal_theta, cost, grad_at_min, inv_hessian_matrix, fun_calls, grad_calls, warn_flags = optimize.fmin_bfgs(
    cost_function_reg,
    initial_theta,
    args=myargs,
    fprime=gradient_reg,
    **options)

print '\nCost at theta found by fmin_bfgs, which is python/scipy equivalent of fminunc:', cost
print '\noptimal theta:'
print optimal_theta
```

```python
def plot_decision_boundary(theta, X, y):
    """
    Plots data points X and y with decision boundary defined by theta. + for positive examples, o for negatives.
    X is assumed to be either:
    1) Mx3 matrix, where the first column is an all-ones column for the intercept.
    2) MxN, N>3 matrix, where the first column is all-ones
    """
    # Clear and close all figures
    clear_output()  #iPython Notebook command
    plt.clf()
    
    plot_data(X[:, 1:3], y)
    if X.shape[1] <= 3:
        plot_x = np.r_[X[:,2].min()-8,  X[:,2].max()+8]   # r_ concatenates scalars into 1-D array
        plot_y = (-1/theta[2]) * (theta[1]*plot_x + theta[0])

        plt.plot(plot_x, plot_y, '-b', linewidth=1)
    else:
    # this part of the code I adapted from Nonnormalizable:
    # https://github.com/Nonnormalizable/NgMachineLearningPython/blob/6f114bf4b9e1c9229b8196b0f6ee1858da576d0a/ex2/helperFunctions.py
        z = np.zeros([50, 50])
        uu = np.linspace(-1.0, 1.5, 50)
        vv = np.linspace(-1.0, 1.5, 50)
        for i, u in enumerate(uu):
            for j, v in enumerate(vv):
                z[i, j] = np.dot(map_feature(np.array([[u]]), np.array([[v]])),
                                        theta)[0]
        plt.contour(uu, vv, z.T, [0], colors='g')
```

```python
# Plot decision boundary

plot_decision_boundary(optimal_theta, X, y)

plt.axis([-.8, 1.2, -.8, 1.2])
plt.xticks(np.arange(-.8, 1.3, .2))
plt.yticks(np.arange(-.8, 1.3, .2))
plt.xlabel('Microchip Test 1')
plt.ylabel('Microchip Test 2')
plt.legend(('y=1', 'y=0'), loc='upper right', numpoints=1)
plt.title('Decision boundary for lambda = ' + str(lmda), fontsize=16)
```

```python
# Computer accuracy on our training set

def predict(theta, X):
    h_theta = h_of_theta(theta, X)
    return np.round(h_theta)

p = predict(optimal_theta, X)

print '\nTraining Accuracy: ', np.mean((p == y) * 100)
```

```python
# Example of overfitting with NO regularization, lambda = 0.00001

lmda = 0.00001
initial_theta = np.zeros(n)
myargs = (X, y, lmda)
options = {'full_output': True, 'maxiter': 400}

optimal_theta, cost, grad_at_min, inv_hessian_matrix, fun_calls, grad_calls, warn_flags = optimize.fmin_bfgs(
    cost_function_reg, initial_theta, args=myargs, fprime=gradient_reg, **options)

plot_decision_boundary(optimal_theta, X, y)

plt.axis([-.8, 1.2, -.8, 1.2])
plt.xticks(np.arange(-.8, 1.3, .2))
plt.yticks(np.arange(-.8, 1.3, .2))
plt.xlabel('Microchip Test 1')
plt.ylabel('Microchip Test 2')
plt.legend(('y=1', 'y=0'), loc='upper right', numpoints=1)
plt.title('Decision boundary for lambda = ' + str(lmda), fontsize=16)
```
