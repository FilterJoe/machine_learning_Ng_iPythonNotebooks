# Machine Learning ex3 (Ng): Multi-class classification with logistic regression

## Goal

The goal is to handle multi-class classification (more than just binary classification). In this part of ex3, we do this with one-vs-all classification by training multiple regularized logistic regression classifiers. We will see how this is done with a pre-trained neural network in ex3nn.

## Use Case

Logistic Regression is a superverised learning classifier $h_\theta(x)$ to estimate the probabilility of whether a binary output is 1 or 0 based on inputs x. However, many problems require classification beyond 1 or 0. The example in this exercise is recognizing which digit among ten possible digits, but any kind of visual or auditory recognition requires such classification.<br><br>
Among the many classification algorithms, logistic regression is one of the simplest to understand and can be adapted to work for simple problems such as recognizing digits using one-vs-all classification. But it often not the best algorithm. Neural Networks are an example of a more complex algorithm that learns better and therefore does a better job of classification.

## Logistic Regression (vectorized)

```python
# Vectorized logistic regression (with regularization) code is displayed in previous exercise, so to keep notebook uncluttered,
# ex3functions.py has code that was alredy displayed in ex2reg (imported above).

def one_vs_all(X, y, num_labels, lmda, max_iterations):
    m, n = X.shape
    all_theta = np.zeros(shape=(num_labels, n+1), dtype=float)
    X = np.insert(X, 0, 1, axis=1)  # Add intercept term (a column of ones) to x along vertical axis

    initial_theta = np.zeros(n+1)
    options = {'full_output': True, 'maxiter': max_iterations, 'disp': False}   # set disp to True to see messages
                                                                                # that say max iterations exceeded and other info
    for c in range(1, num_labels + 1):  # data (from octave) assumes index starts at 1
        # can use either fmin_cg or fmin_bfgs. If using fmin_cg, remove 'grad_at_min, inv_hessian_matrix' (positions 3, 4)
        optimal_theta, cost, fun_calls, grad_calls, warn_flags = optimize.fmin_cg(
            cost_function_reg,  # see ex3functions.py
            initial_theta,
            args=(X, (y==c).astype(int), lmda),  # note that passing in booleans did NOT work, thus .astype(int)
            fprime=gradient_reg,# see ex3functions.py
            **options)
        all_theta[c-1] = optimal_theta
    return all_theta
```

```python
# train multiple classifiers using logistic regression. Thetas optimized with fmin_cg (nonlinear conjugate gradient algorithm)

start_time = time.time()

lamda = .1
max_iterations = 50
all_theta = one_vs_all(X, y, num_labels, lamda, max_iterations)

print time.time() - start_time, "seconds to optimize with fmin_cg and %s iterations." % max_iterations
```

## Part 3: Predict for One-Vs-All

```python
def predict_one_vs_all(all_theta, X):
    X = np.insert(X, 0, 1, axis=1)  # Add intercept term (a column of ones) to x along vertical axis
    h_theta = sigmoid(X.dot(all_theta.T))
    p = np.argmax(h_theta, axis=1) + 1  # data (from octave) assumes index starts at 1
    return p
```

```python
p = predict_one_vs_all(all_theta, X)
print '\nTraining Accuracy: ', np.mean((p == np.ndarray.flatten(y)) * 100)
```

## Part 4: Accuracy impacts from algorithms, number of iterations

While not part of the coursework, an obvious question is whether accuracy can be improved and at what expense. First try changing number of iterations, then changing algorithm.

These timings will vary by hardware. These results are from an Intel i5-4330M Thinkpad running Linux Mint 17.

```python
# increase iterations to 100

start_time = time.time()

lamda = .1
max_iterations = 100
all_theta = one_vs_all(X, y, num_labels, lamda, max_iterations)

print time.time() - start_time, "seconds to optimize with fmin_cg and %s iterations." % max_iterations

p = predict_one_vs_all(all_theta, X)
print '\nTraining Accuracy: ', np.mean((p == np.ndarray.flatten(y)) * 100)
```

```python
# change from fmin_cg algorithm to fmin_bfgs, a quasi-Newton method of Broyden, Fletcher, Goldfarb, and Shanno (BFGS)
# keep iterations at 100

def one_vs_all(X, y, num_labels, lmda, max_iterations):
    m, n = X.shape
    all_theta = np.zeros(shape=(num_labels, n+1), dtype=float)
    X = np.insert(X, 0, 1, axis=1)  # Add intercept term (a column of ones) to x along vertical axis

    initial_theta = np.zeros(n+1)
    options = {'full_output': True, 'maxiter': max_iterations, 'disp': False}   # set disp to True to see messages
                                                                                # that say max iterations exceeded and other info
    for c in range(1, num_labels + 1):  # data (from octave) assumes index starts at 1
        # can use either fmin_cg or fmin_bfgs. If using fmin_cg, remove 'grad_at_min, inv_hessian_matrix' (positions 3, 4)
        optimal_theta, cost, grad_at_min, inv_hessian_matrix, fun_calls, grad_calls, warn_flags = optimize.fmin_bfgs(
            cost_function_reg,  # see ex3functions.py
            initial_theta,
            args=(X, (y==c).astype(int), lmda),  # note that passing in booleans did NOT work, thus .astype(int)
            fprime=gradient_reg,# see ex3functions.py
            **options)
        all_theta[c-1] = optimal_theta
    return all_theta

start_time = time.time()

lamda = .1
max_iterations = 100
all_theta = one_vs_all(X, y, num_labels, lamda, max_iterations)

print time.time() - start_time, "seconds to optimize with fmin_bfgs and %s iterations." % max_iterations

p = predict_one_vs_all(all_theta, X)
print '\nTraining Accuracy: ', np.mean((p == np.ndarray.flatten(y)) * 100)
```

Increasing the number of iterations helps, obviously. For this particular problem, BFGS requires more run time and a higher number of iterations to obtain the same level of accuracy as CG (conjugate gradient). This is true over a wide variety of number of iterations. Clearly the scipy implementation of CG  works better for this particular problem than BFGS.
